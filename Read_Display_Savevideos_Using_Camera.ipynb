```python
import cv2

#In cv2 package their is a class VideoCapture, we'll take this class and create an object named "cap" of this class and pass file name with extension .avi or .mp4 if you want to use already stored video OR if you want to use your device webcam try passing device index: 0 or -1.
#cap = cv2.VideoCapture('myfile.avi/mp4')
#cap = cv2.VideoCapture(0)

cap = cv2.VideoCapture(0)

#we need to create a while loop in order to capture the frame continuously.
while(True):
    ret, frame = cap.read()
    
    #to show the frame
    cv2.imshow('frame',frame)
    if cv2.waitKey(1) == ord('q'):
        break
    
cap.release() # to release cap variable.
cv2.destroyAllWindows() # to destroy the window.
```

```python
# To change frame from coloured video to gray.

import cv2

cap = cv2.VideoCapture(0)

while(True):
    ret, frame = cap.read()
    
    # to change video to gray scale image
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
    # first argument takes the source in my case source is frame which we are capturing using 'cap.read()' method.
    # second argument is conversion what we want to do
    # this gray variable we can pass into imshow() method as input
    
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```

```python
# some properties of 'cap'

import cv2

cap = cv2.VideoCapture(0)

print(cap.isOpened())  #this cap.isOpened() function will return a boolean value it will return True if we enter a valid file name or device. Else it will return False if we enter invalid file name or device. 

while(cap.isOpened()): 
    ret, frame = cap.read()
    
    print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) # it will return width of the frame
    
    print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) ## it will return height of the frame

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) 
    
    cv2.imshow('frame',gray)
    if cv2.waitKey(1) == ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows()
```

```python
# how to save the image which we have captured from our webcam or another default camera.

import cv2

cap =cv2.VideoCapture(0)

# to get fourcc code we are going to use a class called VideoWriter_fourcc.
fourcc = cv2.VideoWriter_fourcc(*'XVID')

# For creating the capture we have used the VideoCapture class and for saving the vido we are going to use the VideoWriter class.
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640,480))
# first argument is name of our output file
# second argument is a fourcc code, now fourcc code is a 4 byte code which is used to specify the video codec
# to know more about fourcc code visit the website: https://www.fourcc.org/codecs.php
# third argument is number of frames per second.
# fourth argument is size

while(cap.isOpened()): 
    ret, frame = cap.read()
    
    if ret == True:
        
        # we can write this frame into a file using a method called out.write()
        out.write(frame)
        
        cv2.imshow('frame',frame)
        if cv2.waitKey(1) == ord('q'):
            break
    else:
        break

cap.release()

# relase all the resources using out.release
out.release()

cv2.destroyAllWindows()
```
